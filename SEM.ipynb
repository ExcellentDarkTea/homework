```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import semopy

df_join = pd.read_csv('join_data_v2.csv')
```

```python
#Use the =~ symbol for a definition of a latent variable as follows: 
#a_latent_variable =~ measured_var_1 + measured_var_2

#Use the ~ symbol for a regression from one latent variable on one or multiple other latent variables as follows:
#latent_variable_1 ~ latent_variable_2 + latent_variable_3

#Use the ~~ symbol if you expect some of the measured variables to have a correlation that is not represented in a latent variable (not necessary for example) as follows:
#measured_var_1 ~~ measured_var_2
model_spec = """
  # measurement model
    load =~ type
    
    Physiological_workload_markers =~ load + type
    HR =~ Physiological_workload_markers
    GSR =~ Physiological_workload_markers
    RR =~ Physiological_workload_markers
    TEMP =~ Physiological_workload_markers
    
    HR ~ hr_min + hr_max + hr_q25 + hr_q50 + hr_q75
    GSR ~  gsr_min + gsr_max_min
    TEMP ~ temp_mean + temp_min + temp_max + temp_q25 + temp_q50 + temp_q75 + temp_max_min
    RR ~ rr_min + rr_max
    
    Personal_trait =~ load
    Demographyc =~ Personal_trait + type
    Psychological_traits =~ Personal_trait + type
    
    Demographyc ~ age + sex_male
    honesty =~ Psychological_traits 
    honesty ~ fairness + modesty + greed_avoidance + sincerity
    
    extraversion =~ Psychological_traits 
    extraversion ~ social_boldness + sociability + liveliness
    
    agreeableness =~ Psychological_traits 
    agreeableness ~ flexibility + patience + gentleness
    
    conscientiousness =~ Psychological_traits 
    conscientiousness ~ prudence + diligence
  
    
    Psychological_traits ~ dependence + openness
"""
# Instantiate the model
model6 = semopy.Model(model_spec)

# Fit the model using the data
model6.fit(df_join)

# Show the results using the inspect method
model6.inspect()

g1 = semopy.semplot(model6, "WWWimage_v2_1.png")

# Open the PNG image file
img = Image.open('WWWimage_v2_1.png')

# Display the image
img.show()
```

```python
import numpy as np
from sklearn.covariance import GraphicalLasso
import networkx as nx
import matplotlib.pyplot as plt

# Instantiate graphical LASSO with alpha (penalty term) value
alpha = 0.5
gl = GraphicalLasso(alpha=alpha)

# Fit the model to the data
gl.fit(df_join)

# Retrieve the estimated precision matrix (inverse of the covariance matrix)
precision_matrix = gl.precision_

# Convert the precision matrix to an adjacency matrix
adjacency_matrix = np.abs(precision_matrix) > 0

# Create a networkx graph from the adjacency matrix
G = nx.from_numpy_array(adjacency_matrix)

# Plot the graph
plt.figure(figsize=(8, 6))
nx.draw_networkx(G, with_labels=True, node_color='lightblue', node_size=300, font_size=12, font_color='black', edge_color='gray', width=0.5)
plt.title('Graphical LASSO Estimated Network')
plt.show()
```

```python
# Define edge widths based on the strengths of the connections
edge_widths = [precision_matrix[i, j]/10 for i, j in G.edges()]

# Plot the graph with thicker edges for stronger connections
plt.figure(figsize=(8, 6))
nx.draw_networkx(G, with_labels=True, node_color='lightblue', node_size=300, font_size=12, font_color='black', edge_color='gray', width=edge_widths)
plt.title('Graphical LASSO Estimated Network with Stronger Connections')
plt.show()
```

```python
import seaborn as sns

corr_matrix = np.clip(precision_matrix, 0, 1)
    
# Apply Fisher's Z transformation to normalize the correlation coefficients
z_corr_matrix = np.arctanh(corr_matrix)
np.fill_diagonal(z_corr_matrix, 0)
sns.heatmap(z_corr_matrix)
```

```python
# Convert the precision matrix to an adjacency matrix
adjacency_matrix2 = np.abs(z_corr_matrix) > 0

# Create a networkx graph from the adjacency matrix
G2 = nx.from_numpy_array(adjacency_matrix2)
```

```python
column_names = df_join.columns.tolist()
node_labels = {i: column_names[i] for i in range(len(column_names))}
```

```python
# Define edge widths based on the strengths of the connections
edge_widths = [precision_matrix[i, j]/10 for i, j in G.edges()]

# Plot the graph with thicker edges for stronger connections
plt.figure(figsize=(150, 50))
pos = nx.spring_layout(G2, seed=42)
nx.draw(G2, pos, labels=node_labels, node_size=1000, font_size=12,
        edge_color='gray', width=edge_widths, edge_cmap=plt.cm.Reds, with_labels=True)
plt.title('Graphical LASSO Estimated Network with Stronger Connections')
plt.show()
```
